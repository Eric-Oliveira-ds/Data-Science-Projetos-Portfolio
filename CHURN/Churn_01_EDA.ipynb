# **O quê é Churn ?**
 
#### **Churn, ou taxa de churn, é uma métrica relacionada à rotatividade de clientes de um negócio. Dessa forma, está diretamente relacionado com os índices de evasão de clientes e cancelamentos de contratos.**
 
#### **Se você tem perdido clientes, a taxa de churn possibilita o diagnóstico. Essa é uma métrica cujo resultado ideal é o mais baixo possível, mas nem por isso ela deixa de ser importante. Ele permite a você identificar se a saúde do negócio está bem ou se inspira cuidados.**
#### **Afinal, qual empresa sobrevive sem clientes, não é mesmo? Eles são mesmo fundamentais para o seu faturamento. Ao identificar a taxa de churn, você descobre se está mais cancelando contratos ou fechando novos. É claro que alguma rotatividade é sempre esperada, seja por uma mudança nas necessidades do cliente ou por iniciativa da concorrência. Ou seja, nem sempre são as suas ações que determinam o sucesso ou o fracasso de uma estratégia de retenção e consequência redução de churn.** 
 
#### **Mas isso não significa se acomodar. Compreender essa rotatividade contribui para você gerir os custos de aquisição de clientes, retê-los por mais tempo e garantir que possa crescer de forma sustentável.**
 
**Referência:** [klick pages](https://klickpages.com.br/blog/churn-como-calcular/)

# **Como calcular a taxa de churn ?**
 
#### **A taxa de churn pode ser calculada de diferentes formas.**
 
**Ela pode tanto representar o número total de clientes perdidos, quanto a porcentagem de clientes perdidos em comparação à contagem total de clientes da empresa, além do valor de negócios recorrentes perdidos, entre outras possibilidades.**

**Geralmente, as empresas calculam essa taxa de cancelamento durante um determinado período de tempo, considerando prazos trimestrais ou anos fiscais.
Assim, um dos métodos mais comumente usados ​​para calcular a rotatividade é dividir o número total de clientes que a empresa possuía no início do período analisado pelo número de clientes perdidos durante o mesmo período.**

**Quer entender melhor a partir de um exemplo?**

**Então, vamos lá!**

**Vamos supor que cinco dos seus clientes   cancelaram o serviço no primeiro trimestre do ano e que você tinha originalmente 100 clientes durante esse período.**

**Assim, seu churn rate seria de 5%.**

**Veja: 5 ÷ 100 = 0,05 x 100% = 5%**

**Mas será que esses 5% são uma taxa boa, ruim ou aceitável ? Depende da amplitude do número de clientes e do seu modelo de negócio.**

**Existe taxa de churn ideal ? Algumas empresas consideram uma taxa aceitável entre 5% e 7%.**

# **Preparando os dados para EDA**

**Vou transformar as colunas : HasCrCard, IsActiveMember, NumOfProducts, Exited em categóricas para facilitar a interpretação para o time de finanças.**
 
**De acordo ao dicionário de dados é possível sempre a interpretabilidade de cada coluna da tabela.**

```python
# Removerei temporariamente colunas irrelevantes para a análise aqui 
 
df1 = df.drop(["RowNumber","Surname"],axis=1)
```

```python
# Transformando as colunas em categorias binarias de texto 
 
# Criei uma cópia do dataframe para trabalhar sem nesscecitar de mudar o original 
df_cat = df1.copy()
 
# Onde houver 1 substituí por (tem cartão) e onde houver 0 substituí por  (não tem cartão)
df_cat["HasCrCard"] = df_cat["HasCrCard"].apply(lambda x: "Tem cartão" if x == 1 else "Não tem cartão")
 
# Onde houver 1 recebe (membro ativo), onde houver 0 (membro inativo)
df_cat["IsActiveMember"] = df_cat["IsActiveMember"].apply(lambda x: "membro ativo" if x == 1 else "membro inativo")
 
# Onde houver 1 recebi  (Saiu), onde houver 0 (Não saiu)
df_cat["Exited"] = df_cat["Exited"].apply(lambda x: "Saiu" if x == 1 else "Não saiu")
 
# Transformei a coluna número de produtos objeto
df_cat["NumOfProducts"] = df_cat["NumOfProducts"].astype("str")
 
# Mudei os nomes dos gêneros (Female -> Mulher) e (Male -> Homem)
df_cat["Gender"] = df_cat["Gender"].replace(["Female","Male"],["Mulher","Homem"]) #("Male","Homem")
```

```python
# Exibindo o dataframe para verificar as mudanças 
df_cat.head()
```

**1 - Quantos clientes já deixaram o banco, e quantos permaneceram ?**

**3 - Há relação entre a estabilidade no emprego e a saída do cliente ?**

```python
# filtrei apenas as colunas onde pertencem a clientes que saíram do banco e agreguei a moda de estabilidade no emprego
# A moda é o valor mais frequente em relação ao índice de estabilidade no emprego que aparece
df_cat[df_cat["Exited"] == "Saiu"].agg({"Tenure":"mode"})
```

```python
# Gráfico de barras para comprovar a frequência do índice de estabilidade no emprego para clientes que saem do banco 
 
# Filtrei apenas as linhas que pertencem a clientes que saíram, para fazer uma abordagem conforme a pergunta do CFO
df_tenure = df1[df1["Exited"] == 1 ]
 
# Agrupei o índice de estabilidade e somei por clientes em churn para cada valor de 0 a 10 na métrica de estabilidade.
df_tenure.groupby("Tenure").agg({"Exited":"sum"}).plot(kind="barh",
                                                  figsize=(16,8), 
                                                  title="Frequência de churn por índice de estabilidade no emprego");
```
