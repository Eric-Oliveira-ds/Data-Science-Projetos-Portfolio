**Pré-processamento dos dados**

```python
# Selecionei as variáveis categoricas que irei usar na modelagem para transformação em numéricas para que o modelo possa aprender,  pois o algoritmos só compreende números 
df_cat = df[["Geography","Gender"]]
```

```python
# Usando função para transformar as variáveis categóricas em numéricas 
df_cat = pd.get_dummies(df_cat)
df_cat.head()
```

```python
# pegando só as colunas com dados numéricos , exceto a variável alvo "Exited"
df_num = df[["CreditScore","Age","Tenure","Balance","NumOfProducts","HasCrCard","IsActiveMember","EstimatedSalary"]]                 
df_num.head()
```

```python
# juntando as bases em um único dataframe 
X = pd.concat([df_cat, df_num], axis=1)
```

```python
# Variáveis explicativas, seu comprimento 
X.shape
```

```python
# Variável alvo, seu comprimento 
y = df["Exited"]
y.shape
```

```python
# Verificando que as classes são desbalanceadas 
y.value_counts()
```

**Balanceamento de classes**

```python
# Usei a técnica de sobreamostragem de dados SMOTE, que seleciona os k vizinhos mais próximos da classe majoritária e cria uma reta conforme cálculo de proximidade para classe minoritária, e gera amostras sintéticas na classe minoritária para seu balanceamento
oversample = SMOTE()
X_over, y_over = oversample.fit_resample(X, y)
```

```python
# Verificando se as classes estão no mesmo tamanho, "balanceadas"
pd.Series(y_over).value_counts ()
```

```python
# tamanho final dos dados depois da amostragem 
X_over.shape, y_over.shape
```

**Separação dos dados entre treino e teste**

```python
# Separação dos dados entre treino e validação 
X_train, X_valid, y_train, y_valid = train_test_split(X_over, y_over, test_size=0.2, stratify=y_over ,random_state=42)
```

```python
# Verificando tamanho dados de treino 
X_train.shape, y_train.shape
```

```python
# Verificando tamanho dados de validação 
X_valid.shape, y_valid.shape
```

**Escalonamento dos dados**

```python
# Verificando as escalas dos dados, onde existirem dados com valores dispersos, e variados irei escalonar para facilitar o aprendizado do modelo
pd.DataFrame(X_train)
```

```python
 # define o  scaler
scaler = MinMaxScaler()
 
# fit nos dados de treino , trás todos os dados para um intervalo entre (0 e 1)
scaler.fit(X_train)
 
# scaler transformação nos dados de treino 
X_train = scaler.transform(X_train)
 
# scaler transformação nos dados de validação 
X_valid = scaler.transform(X_valid)
```

```python
pd.DataFrame(X_train)
```

**Modelos de Machine Learning**

```python
# Validação cruzada estratificada com 10 folds para cada algoritmo
 
for model in [
    LogisticRegression,
    DecisionTreeClassifier,
    ExtraTreeClassifier,
    RandomForestClassifier,
    SVC,
    GaussianNB,
    KNeighborsClassifier,
    XGBClassifier,
    LGBMClassifier]:
        
        # Pega a lista de modelos como objeto 
        clfs = model()
    
        # Validação cruzada estratificada
        skfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
        
        # Cálculo da validação pela métrica ROC_AUC, (nos dados de validação) 
        auc = cross_val_score(clfs, X_valid, y_valid, scoring="roc_auc", cv=skfold, n_jobs=-1)
 
        # Treina os modelos 
        clfs.fit(X_train, y_train)
        
        # Faz predições
        yhat = clfs.predict(X_valid)
 
        # Exibe as métricas 
        print(f"{model.__name__} AUC_CV: "f"{auc.mean():.3f} STD: {auc.std():.2f}",end="\n\n")
        print(f"{classification_report(y_valid, yhat)}")
        print(f"{confusion_matrix(y_valid, yhat)}")
        print("....................................................................")
```
