# **Dicionrio de dados:**
 
**Atributos para os conjuntos de dados student-mat.csv (curso de matemtica) e student-por.csv (curso de lngua portuguesa):**
 
**1 - school - escola do aluno (binrio: 'GP' - Gabriel Pereira ou 'MS' - Mousinho da Silveira)**
 
**2 - sex - sexo do aluno (binrio: 'F' - feminino ou 'M' - masculino)**
 
**3 - age - idade do aluno (numrico: de 15 a 22)**
 
**4 - address - tipo de endereo residencial do aluno (binrio: 'U' - urbano ou 'R '- rural)**
 
**5 - famsize - tamanho da famlia (binrio:' LE3 '- menor ou igual a 3 ou' GT3 '- maior que 3)**
 
**6 - Pstatus - status de coabitao dos pais (binrio:' T '- morando junto ou' A '-  parte)**
 
**7 - Medu - escolaridade da me (numrico: 0 - nenhuma, 1 - ensino fundamental (4 srie), 2 - 5 a 9 srie, 3 - ensino mdio ou 4 - ensino superior)**
 
**8 - Fedu - escolaridade do pai (numrico: 0 - nenhuma, 1 - ensino fundamental (4 ano), 2 - 5 ao 9 ano, 3 - ensino mdio ou 4 - ensino superior)**
 
**9 - Mjob - trabalho da me ( nominal: 'professor', 'cuidados de sade' relacionados, 'servios' civis (por exemplo, administrativos ou policiais), 'em_casa' ou 'outro')**
 
**10 - Fjob - trabalho do pai (nominal: 'professor', 'cuidados de sade' relacionados, civil 'servios' (por exemplo, administrativo ou policial), 'em_ casa' ou 'outro')**
 
**11 - reason - razo para escolher esta escola (nominal: perto de 'casa', escola 'reputao', 'curso' preferncia ou 'outro')**
 
**12 - guardian - tutor do aluno (nominal: 'me', 'pai' ou 'outro' )**
 
**13 - travel time  - tempo de viagem de casa para a escola (numrico: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. A 1 hora, ou 4 -> 1 hora)**
 
**14 - studytime - tempo de estudo semanal (numrico: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas ou 4 -> 10 horas)**
 
**15 - failures - nmero de reprovaes anteriores nas aulas (numrico: n se 1 <= n <3, seno 4)**
 
**16 - schoolsup - suporte educacional extra (binrio: sim ou no)**
 
**17 - famsup - suporte educacional da famlia (binrio: sim ou no)**
 
**18 - paid - aulas extras pagas dentro da matria do curso (matemtica ou portugus) (binrio: sim ou no)**
 
**19 - activities - atividades extracurriculares (binrio: sim ou no)**
 
**20 - nursery - cursou creche (binrio: sim ou no)**
 
**21 - higher - deseja cursar o ensino superior (binrio: sim ou no)**
 
**22 - internet - acesso  internet em casa (binrio: sim ou no)**
 
**23 - romantic - com um relacionamento romntico (binrio: sim ou no)**
 
**24 - famrel - qualidade das relaes familiares (numrico: de 1 - muito ruim a 5 - excelente)**
 
**25 - freetime - tempo livre depois da escola (numrico: de 1 - muito baixo a 5 - muito alto)**
 
**26 - goout - sair com os amigos (numrico : de 1 - muito baixo a 5 - muito alto)**
 
**27 - Dalc - consumo de lcool durante o trabalho (numrico: de 1 - muito baixo a 5 - muito alto)**
 
**28 - Walc - consumo de lcool no fim de semana (numrico: de 1 - muito baixo a 5 - muito alto)**
 
**29 - health - estado de sade atual (numrico: de 1 - muito ruim a 5 - muito bom)**
 
**30 - absences - nmero de faltas na escola (numrico: de 0 a 93)**
 
**Essas notas esto relacionadas com a disciplina, Matemtica ou Portugus :**
 
**31 - G1 - nota do primeiro perodo (numrico: de 0 a 20)**
 
**32 - G2 - nota do segundo perodo (numrico: de 0 a 20)**
 
**33 - G3 - nota final (numrico: de 0 a 20, meta de sada)**
